```python
#Create a 3 different US state-wise visualization that color-maps the quantity of any information you collected 
df_time = pd.read_csv("F:/PA2/Input/us.csv")
import altair as alt
from vega_datasets import data
states = alt.topo_feature(data.us_10m.url, 'states')
variable_list = ['Murder', 'Assault', 'Rape']
alt.Chart(states).mark_geoshape().encode(
 alt.Color(alt.repeat('row'), type='quantitative')
).transform_lookup(
 lookup='id',
 from_=alt.LookupData(df_time, 'id', variable_list)
).properties(
 width=700,
 height=500
).project(
 type='albersUsa'
).repeat(
 row=variable_list
).resolve_scale(
 color='independent'
)
```
