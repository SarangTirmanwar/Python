```python
import altair as alt
#creating a variable small and adding first 16 rows in that variable
small=df.head(16)
small
```

```python
#1Univariate and Multivariate Visualizations
# A.i) A bar graph that shows categorical count 
alt.Chart(small).mark_bar(size=40).encode(
 alt.X("name:N"),
 alt.Y("average(budget)")).properties(
    width=700,
    height=500
 )

       
```

```python
#ii. A histogram graph that shows a quantitative distribution
from altair import datum
alt.Chart(small).mark_bar().encode(
 alt.X("score:Q", bin=True),
 y='count()',
).transform_filter(
 (datum.rating== 'PG')
)

```

```python
#B.i) One box plot that compares more than two groups
alt.Chart(small).mark_boxplot().encode(
alt.X("budget:Q"),
alt.Y("gross:Q")
)
```

```python
#ii)One strip plot that compares more than two groups
stripplot = alt.Chart(small, width=70).mark_circle(size=70).encode(
    alt.X(
          'jitter:Q',
           title=None,
           axis=alt.Axis(values=[0],
           ticks=True,
           grid=False,
           labels=False), scale=alt.Scale()),
     alt.Y('score:Q'),
     color=alt.Color('year:Q', legend=None),
     column=alt.Column(
               'year:Q',
                header=alt.Header(labelAngle=0, titleOrient='top', labelOrient='bottom', labelAlign='center',labelPadding=20),
 ),
).transform_calculate(
 jitter='sqrt(-2*log(random()))*cos(2*PI*random())'
).properties(width=70)
stripplot
```

```python
#C.i) A scatter plot that shows how two quantitative columns
alt.Chart(small).mark_circle(size=60).encode(
 alt.X("budget:Q"),
 alt.Y("gross:Q"),
 color='score:Q',
 size='score:Q'
).properties(width=700, height=450)
```

```python
#ii) Draw a regression line for the scatter plot 
Base=alt.Chart(small).mark_circle(size=60).encode(
 alt.X("budget:Q"),
 alt.Y("gross:Q"),
 color='score:O',
 size='score:O'
).properties(width=700, height=450)
Line = Base.transform_regression('budget', 'gross', method='poly').mark_line(size=5)
alt.layer(Base, Line)
```
